# Largest Component

Take a graph:<br>We want to return the largest component. I.e. largest no. of nodes along the traversal.<br>In this case, should return 4.

## Pseudo-Code

Need:
* Iterative code to hop to different components
* Depth first traversal variation to find the *size* of a connected component.
* And also, as standard, a set to keep track of visited nodes.

list out nodes.keys() to for the list of nodes to iterate through.<br>
Traversal through while marking nodes as visited.<br>
And want to count every node as we traverse through it.<br>
0 first, all the way through, then 2 (skip 1 bc visited).

Complexity is all of those seen before: time: O(e), and space: O(n)

## Code

```python
def explore_size(graph, current, visited):
    if current in visited:
        return 0
    visited.add(current)
    size=1
    for neighbour in graph[current]:
        size += explore_size(graph, neighbour, visited)
    return size

def largest_component(graph):
    visited = set()
    max_size = 0
    for node in graph.keys():
        size = explore_size(graph, node, visited)
        if size > max_size:
            max_size = size
    return max_size
```

**Testing the Code**
