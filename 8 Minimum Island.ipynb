# 8 Minimum Island

Grid-graphs like the previous exercise, how can I answer what is the smallest island?<br>
<br>By appearance, we know we should return 2.<br>This is a spin-off of the previous problem with a **counts** as we traverse each.

## Pseudo-code

Will need:
* Nested code.
* Depth-first traversal code to explore a single island.
* Visited set to prevent loops.
* Island count variable. This will probably re-use recursive code from 5-largest-component.

<p>When going through, consider the first island as min-size. After, testing whether is min size.</p>

Complexity:<br>
Time: $O(rc)$
Space: $O(rc)$

## Code

```python
def island_size(grid, row, col, visited):
    """See 7-island-count for more details on what's going on here."""
    row_in_bounds = row >= 0 and row < (len(grid))
    col_in_bounds = col >= 0 and col < (len(grid[0]))
    
    if (not row_in_bounds or  not col_in_bounds):
        return 0 #represents is out of bounds; doesn't contribute to the count of the size.
    
    if grid[row][col] =='W':
        return 0
    
    pos = (row,col)
    if pos in visited:
        return 0
    visited.add(pos)
    
    #size represents current position
    size = 1
    
    size += island_size(grid, row-1, col, visited)
    size += island_size(grid, row+1, col, visited)
    size += island_size(grid, row, col-1, visited)
    size += island_size(grid, row, col+1, visited)
    
    return size

def minimum_island(grid):
    visited = set()
    min_size = float('inf')
    for row in range(0,len(grid)):
        for col in range(0,len(grid[0])):
            size = island_size(grid, row, col, visited)
            if (size < min_size and size > 0):
                min_size = size
    return min_size
```

**Testing the Code**

```python
grid = [
    ['W','L','W','W','W'],
    ['W','L','L','W','W'],
    ['W','W','W','L','W'],
    ['W','W','L','L','W'],
    ['W','L','W','L','L'],
    ['L','L','W','W','W']
]
```

```python
minimum_island(grid=grid)
```
